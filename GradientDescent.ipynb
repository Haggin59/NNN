##Gradient Descent


```python
import numpy as np
import matplotlib.pyplot as plt

from IPython import display
display.set_matplotlib_formats('svg')
```

f(x) = 3x^2 -3x + 4  
f'(x) = 6x -3

```python
#f(x)
def fx(x):
  return 3*x**2 - 3*x + 4

#f'(x)
def dfx(x):
  return 6*x - 3
```

```python
x = np.linspace(-2,2,2001)

plt.plot(x,fx(x),'r',x,dfx(x),'b')
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['f(x)','df(x)'])
```

Running GD

```python
local_min = np.random.choice(x,1)    #initializing random local minimum
print("Random :",local_min)

learning_rate = 0.1
epoch = 100

for i in range(epoch):
  grad = dfx(local_min)                        #calculating slope(gradient) at local min
  local_min = local_min - learning_rate*grad   #going downhill, if grad is -ve: local_min increases, if +ve decreses

print(local_min)
print(f"dfx({local_min[0]}) :",dfx(local_min[0]))
print(f"f({local_min[0]}) :",fx(local_min[0]))
```

```python

plt.plot(x,fx(x),'r',x,dfx(x),'b')
plt.plot(local_min, fx(local_min), 'ro')
plt.plot(local_min, dfx(local_min), 'ro')
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['f(x)','df(x)'])
```

Change in grad and local_min

```python
local_min = np.random.choice(x,1)  
print("Random :",local_min)

learning_rate = 0.01
epoch = 100

GD_params = np.zeros((epoch,2))

for i in range(epoch):
  grad = dfx(local_min)                        
  local_min = local_min - learning_rate*grad 
  GD_params[i,:] = local_min,grad
```

```python
dfx(-0.43)
```

```python
GD_params[0:5]
```

Plotting change in local_minima & gradient through epochs

```python

plt.subplot(1,2,1)
plt.plot(GD_params[:,0],'-b.')
plt.xlabel('iterations')
plt.ylabel('local minima')


plt.subplot(1,2,2)
plt.plot(GD_params[:,1],'-r.')
plt.xlabel('iterations')
plt.ylabel('gradient')
```

```python
fig,ax = plt.subplots(1,2,figsize=(12,4))
ax[0].plot(GD_params[:,0],'-o')
ax[0].set_xlabel('Iterations')
ax[0].set_ylabel('local minima')
ax[0].set_title(f'Estimated Loacal Minima : {local_min[0]:.4f}')

ax[1].plot(GD_params[:,1],'-o')
ax[1].set_xlabel('Iterations')
ax[1].set_ylabel('gradient')
```
